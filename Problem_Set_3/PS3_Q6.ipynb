```python
import numpy as np
from astropy.io import fits
from matplotlib import pyplot as plt
from scipy import fft
import time
```

# Loading data

```python
hdul = fits.open('advact_tt_patch.fits')
map = hdul[0].data
hdul.close()
map = np.asarray(map,dtype='float')
print('read map')
```

# Finding the new cluster on map

```python
plt.imshow(map, vmin = -100.0, vmax = 100)
#plt.xlim(2970, 3030)
plt.xlim(3170, 3230)
plt.ylim(1980, 2040)
plt.colorbar()
plt.title('New Cluster')
plt.show()
```

# Calculating the RA and Dec

According to the NED, the RA and DEc for ABELL2813 are :

$RA = 00h43m24.4s$ and
$Dec = -20d37m17s$

Now, regarding the fact that this new cluster is roughly 200 pixels to the right and 10 pixels to the up from the ABELL2813, and each pixel is 30 arcseconds, the RA is expected in decrease by 600 and the Dec to increase by 300 in value.

Thus, the expected coordinates of the new cluster is:

$RA = 00h33m24.4s$
$Dec = -20d32m17s$
